```python
import matplotlib.pyplot as plt
import numpy as np

with open("WdMadrigal16-07.lnd", "rb") as file:
    bs = file.read()
offset = 1200
ints = np.frombuffer(bs[offset*4:len(bs)//4*4], dtype=np.int32)
sidelen = int(np.sqrt(len(ints)))
ints = ints[:sidelen*sidelen]
map = np.reshape(ints, (sidelen, sidelen))
plt.imshow(map)
```
